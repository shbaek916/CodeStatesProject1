# Number of Buildings 추출

```python
import pandas as pd
```

```python
building = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Codestates Project 1/raw_data/train10000.csv').fillna('NaN')
```

```python
building[(building.game_id == 182)&(building.player==0)&(building.event_contents.str.contains('Build'))]
```

```python
players = [0,1]
build = []
start = 0
for num in [10000, 20000, 30000, 38871]:
  data = pd.read_csv(f'/content/drive/MyDrive/Colab Notebooks/Codestates Project 1/raw_data/train{num}.csv').fillna('NaN')
  for i in range(start, num+1):
    game = data[data.game_id == i]
    tmp = [0, 0]
    filtered = game[(game.event_contents.str.contains('Build'))&(~game.event_contents.str.contains("Hellion|SiegeTank|WidowMine|Thor"))]
    filter_cancel = game[game.event_contents.str.contains('CancelBuilding|CancelTerranBuilding')]
    for player in players:
      player_buildings = filtered[filtered.player == player]
      player_cancel = filter_cancel[filter_cancel.player == player]
      tmp[player] = len(player_buildings) - len(player_cancel)
    build.append(tmp)
  if num == 10000:
    start = 10001
  else:
    start += 10000
  del data
build
```

```python
build = pd.DataFrame(build)
build.to_csv('/content/drive/MyDrive/Colab Notebooks/Codestates Project 1/raw_data/build.csv',sep=',',index=False)
```

# Attack 명령 수 추출


```python
train[train.event_contents.str.contains('Attack')]
```

```python
players = [0,1]
attack = []
start = 0
for num in [10000, 20000, 30000, 38871]:
  data = pd.read_csv(f'/content/drive/MyDrive/Colab Notebooks/Codestates Project 1/raw_data/train{num}.csv').fillna('NaN')
  for i in range(start, num+1):
    game = data[data.game_id == i]
    tmp = [0,0]
    attack_events = game[game.event_contents.str.contains('Attack')]
    for player in players:
      player_attack = attack_events[attack_events.player == player]
      tmp[player] = len(player_attack)
    attack.append(tmp)
  if num == 10000:
    start = 10001
  else:
    start += 10000
  del data

attack = pd.DataFrame(attack)
attack
```

# 이벤트 명령 수 추출 및 데이터 축소

```python
practice = train[(train.player ==1)&(train.game_id == 0)]
practice = practice.groupby('player',as_index=False).agg({
    'game_id':'first',
    'winner':'first',
    'time':'last',
    'species':'first',
    'event': '|'.join
    # 'event_contents':'first'
})
practice
```

```python
temp = pd.DataFrame(pd.Series(practice.event[0].split('|')).value_counts()).T
temp
```

```python
start = 0
initial = False
step = None
TF = False
time = []
groupby_list = [
  {'winner':'first',
   'time':'last',
   'species':'first',
   'event': '|'.join},
  {'species':'first',
   'event': '|'.join}
]
for num in [10000, 20000, 30000, 38871]:
# for num in [10000]:
    data = pd.read_csv(f'/content/drive/MyDrive/Colab Notebooks/Codestates Project 1/raw_data/train{num}.csv').fillna('NaN')
    # num=10
    for id in range(start, num+1):
        print(id)
        temp = data[data.game_id == id ]
        for player in [0,1]:
            result = temp[temp.player == player]
            result = result.groupby('player',as_index=False).agg(groupby_list[player])

            event_counts = pd.DataFrame(pd.Series(result.event[0].split('|')).value_counts()).T
            result['event'] = sum(event_counts.iloc[0])
            try:
              result[f'player_{player}_Ability'] = event_counts['Ability']
            except:
              result[f'player_{player}_Ability'] = 0
            if TF == False:
                step = result
                TF = True
            else:
                step = pd.concat([step, result], axis =1)
                TF = False

        if initial == False:
            final = step
            initial = True
        else:
            final = pd.concat([final, step])
            TF = False
    if start == 0:
        start = 10001
    else:
        start += 10000  
final
      
```
