```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
attack = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Codestates Project 1/raw_data/attack.csv')
build = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Codestates Project 1/raw_data/build.csv')
worker10000 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Codestates Project 1/raw_data/worker10000.csv')
worker_rest = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Codestates Project 1/raw_data/worker_rest.csv')
nonworker = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Codestates Project 1/raw_data/nonworker.csv')
worker = pd.concat([worker10000, worker_rest]).reset_index().iloc[:,1:]

train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Codestates Project 1/raw_data/processed_train.csv')

del worker10000
del worker_rest
```

```python
train = train.drop(['player', 'player.1'], axis=1)
train
```

```python
attack.columns = ['player_0_attack','player_1_attack']
build.columns = ['player_0_buildings','player_1_buildings']
worker.columns = ['player_0_worker','player_1_worker']
nonworker.columns = ['player_0_nonworker', 'player_1_nonworker']

train = pd.concat([train, attack, build, worker, nonworker], axis =1)
train.columns
```

```python
train = train[['winner', 'time', 
                 'species', 'event', 'player_0_attack', 'player_0_Ability', 'player_0_buildings', 'player_0_worker', 'player_0_nonworker',
                 'species.1', 'event.1', 'player_1_attack', 'player_1_Ability', 'player_1_buildings', 'player_1_worker', 'player_1_nonworker']]
train
```

```python
train = train.rename(columns={'species':'player_0_species', 'event':'player_0_APM', 
                              'species.1':'player_1_species','event.1':'player_1_APM'})
train
```

```python
train.iloc[1258,:]
```

```python
timeinfloat = []
for i in range(38872):
    t = train.time[i]
    timeinfloat.append( round(t // 1 + t % 1 / 0.6,2) )
train.time = timeinfloat
train
```

```python
train.player_0_APM = round(train.player_0_APM/train.time)
train.player_1_APM = round(train.player_1_APM/train.time)
```
